```python
from sklearn.datasets import load_iris
from sklearn import tree
import numpy as np
import random
```

```python
iris = load_iris()
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
ind = np.arange(150)
```

```python
ind
```

```python
random.shuffle(ind)
iris.data = iris.data[ind]
iris.target = iris.target[ind]
```

```python
iris.target
```

```python
iris_train_x = iris.data[:-50]
iris_test_x = iris.data[-50:]
iris_train_y = iris.target[:-50]
iris_test_y = iris.target[-50:]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(iris_train_x, iris_train_y)
```

```python
predict = clf.predict(iris_test_x)
```

```python
print(predict)
```

```python
numOfCorrectPrediction = (predict == iris_test_y).sum()
numOfDataSamples = len(iris_test_y)
```

```python
print(numOfCorrectPrediction/ float(numOfDataSamples))
```
