```python
class Picture:
    def __init__(self, filepath):
        from PIL.Image import open
        data = open(filepath).getdata()
        self.jpgfile = list(data)
        self.mode = data.mode
        self.size = data.size
        
        self.Color = self.jpgfile
        
        px = []
        for i in range(self.size[0]*self.size[1]):
            px.append([i,self.Color[i]])
            
        self.Pixel = px
        
    def show(self):
        from PIL import Image
        im2 = Image.new(self.mode, self.size)
        im2.putdata(self.jpgfile)
        return im2.show()
    
    def getPixels(self) :
        return self.Pixel
    
    def getPixel(self, xpos, ypos):
        index = ypos*self.size[0]+xpos
        return self.Pixel[index]
    
    def getX(self, pixel):
        width = self.size[0]
        index = pixel[0]
        return index%witdth
    
    def getY(self, pixel):
        width = self.size[0]
        index = pixel[0]
        return width//index  
    
    def getHeight(self):
        return self.size[1]
    
    def getWidth(self):
        return self.size[0]
    
    def getRed(self, pixel) :
        return pixel[1][0]
    
    def getGreen(self, pixel) :
        return pixel[1][1]
    
    def getBlue(self, pixel) :
        return pixel[1][2]
    
    def setRed(self, pixel, value) :
        if(value>255) :
            value = 255
        index = pixel[0]
        
        R = self.Pixel[index][1][0]
        G = self.Pixel[index][1][1]
        B = self.Pixel[index][1][2]
        
        self.Pixel[index][1] = (value,G,B)
        self.Color[index] = (value,G,B)
        
    def setGreen(self, pixel, value) :
        if(value>255) :
            value = 255
            
        index = pixel[0]
        
        R = self.Pixel[index][1][0]
        G = self.Pixel[index][1][1]
        B = self.Pixel[index][1][2]
        
        self.Pixel[index][1] = (R,value,B)
        self.Color[index] = (R,value,B)
        
    def setBlue(self, pixel, value) :  
        if(value>255) :
            value = 255       
        
        index = pixel[0]
        
        R = self.Pixel[index][1][0]
        G = self.Pixel[index][1][1]
        B = self.Pixel[index][1][2]
        
        self.Pixel[index][1] = (R,G,value)
        self.Color[index] = (R,G,value)
    
    def getColor(self,pixel):
        return pixel[1]
    
    def setColor(self, pixel, color):
        index = pixel[0]
        self.Pixel[index][1] = color
        self.Color[index] = color
    
    def makeColor(self, red, *args) :
        if(len(args)==2):
            green = args[0]
            blue = args[1]
            if(red>255):
                red = 255
            if(green>255):
                green = 255
            if(blue>255):
                blue = 255
            color = (red,green,blue)
            return color
        else :
            if(red>255):
                red = 255
            color = (red,red,red)
            return color
  

 
        
class Setting:
    def setMediaPath(self,directory) :
        self.directory = directory

    def getMediaPath(self, filename):
        self.filename = self.directory + '\\' + filename
        return self.filename 

    def makePicture(self, path) :
        picture = Picture(path)
        return picture
    
        
```

```python
def green_ify(path):
    s = Setting()
    pic = s.makePicture(path)
    for p in pic.getPixels():
        red = int(pic.getRed(p) / 2)
        pic.setRed(p,red)
        green = int(pic.getGreen(p) *2)
        pic.setGreen(p,green)
        blue = int(pic.getBlue(p) / 2)
        pic.setBlue(p,blue)
    pic.show()
```
