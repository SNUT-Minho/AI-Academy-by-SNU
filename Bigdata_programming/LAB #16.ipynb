```python
bu = './butterfly1.jpg'
bu = plt.imread(bu)
plt.imshow(bu)
```

```python
fly = np.array(bu)
```

```python
for x in range(fly.shape[0]-1) :
    for y in range(fly.shape[1]-1) :
        here = fly[x][y]
        down = fly[x][y+1]
        right = fly[x+1][y]
        hereLumi = (here[0]/3) + (here[1]/3) + (here[2]/3)
        downLumi = (down[0]/3) + (down[1]/3) + (down[2]/3)
        rightLumi = (right[0]/3) + (right[1]/3) + (right[2]/3)
        if( abs(hereLumi - downLumi) > 10 and abs(hereLumi-rightLumi)>10):
            here[0] = 0
            here[1] = 0
            here[2] = 0
        else:
            here[0] = 255
            here[1] = 255
            here[2] = 255
```

```python
plt.imshow(fly)
```
