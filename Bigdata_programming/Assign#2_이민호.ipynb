```python
class Picture:
    def __init__(self, filepath):
        from PIL.Image import open
        data = open(filepath).getdata()
        self.jpgfile = list(data)
        self.mode = data.mode
        self.size = data.size
        
        Pixels = []
       
        self.color = self.jpgfile
        
        for y in range(self.size[1]):
            for x in range(self.size[0]):
                index = [y,x]
                Color = (self.jpgfile[self.size[0]*y + x])
                Pixels.append([index,Color])
        
        self.pixels = Pixels
        
    def show(self):
        from PIL import Image
        im2 = Image.new(self.mode, self.size)
        im2.putdata(self.jpgfile)
        
        return im2.show()
    
    def getPixels(self) : 
        
        return self.pixels
    
    def getPixel(self, xpos, ypos):
        if(xpos >= self.size[0] or xpos < 0):
            return "xpos is outof bound"
        
        elif (ypos >= self.size[1] or ypos <0):
            return "ypos is outof bound"
        
        index = ypos * self.size[0]+ xpos
        
        return self.pixels[index]
    
    def getX(self, pixel):  
        return pixel[0][1]
    
    def getY(self, pixel): 
        return pixel[0][0]
    
    def getHeight(self):
        return self.size[1]
    
    def getWidth(self):
        return self.size[0]
    
    def getRed(self, pixel) :
        return pixel[1][0]
    
    def getGreen(self, pixel) :
        return pixel[1][1]
    
    def getBlue(self, pixel) :
        return pixel[1][2]
    
    def setRed(self, pixel, value) :
        y = pixel[0][0]
        x = pixel[0][1]
        index = y*self.size[0] + x
        
        r = value
        g = pixel[1][1]
        b = pixel[1][2]
        
        self.pixels[index][1] = (r,g,b)
        self.color[index] = (r,g,b)
           
    def setGreen(self, pixel, value) :
        y = pixel[0][0]
        x = pixel[0][1]
        index = y*self.size[0] + x
        
        r = pixel[1][0]
        g = value
        b = pixel[1][2]
        
        self.pixels[index][1] = (r,g,b)
        self.color[index] = (r,g,b)
        
    def setBlue(self, pixel, value) :  
        y = pixel[0][0]
        x = pixel[0][1]
        index = y*self.size[0] + x
        
        r = pixel[1][0]
        g = pixel[1][1]
        b = value
        
        self.pixels[index][1] = (r,g,b)
        self.color[index] = (r,g,b)
        
    def getColor(self,pixel):
        return pixel[1]          
    
    
    def setColor(self, pixel, color):
        y = pixel[0][0]
        x = pixel[0][1]  
        index = y*self.size[0] + x
          
        self.pixels[index][1] = color
        self.color[index] = color
        
    
    def makeColor(self, red, *args) :
        if(len(args)==2):
            green = args[0]
            blue = args[1]
            color = (red,green,blue)
            return color
        else :
            color = (red,red,red)
            return color
```

```python
class Setting:
    def setMediaPath(self, string):
        import os
        if not string.endswith(os.sep):
            self.PATH = string + os.sep

    def getMediaPath(self, string):
        return self.PATH + string

    def makePicture(self, filepath=""):
        if filepath is "":
            filepath = self.PATH
        return Picture(filepath)      
```

```python
def copy (source,target,target_X,target_Y) :
    x= target_X
    width = source.getWidth() 
    height = source.getHeight()
    
    for source_Y in range(0,height) :
        target_X = x
        for source_X in range(0,width) :
            source_Pixel = source.getPixel(source_X, source_Y)
            target_Pixel = target.getPixel(target_X,target_Y)
            target.setColor(target_Pixel,source.getColor(source_Pixel))
            target_X = target_X +1
        target_Y = target_Y+1   
```

```python
def copy2(source,target,target_X,target_Y) :
    width = source.getWidth()
    height = source.getHeight()
    
    source_Y = 0
    for y in range(0,int(height/4)) :
        source_X = 0
        for x in range(0, int(width/4)) :
            source_Pixel = source.getPixel(source_X, source_Y)
            target_Pixel = target.getPixel(target_X+x,target_Y+y)
            target.setColor(target_Pixel,source.getColor(source_Pixel))
            source_X = source_X +4
        source_Y =source_Y+4       
```

```python
def avengers(thor,cap,hulk,iron,logo,canvas):
   
    se = Setting()
    thor = se.makePicture(thor)
    cap = se.makePicture(cap)
    hulk = se.makePicture(hulk)
    iron = se.makePicture(iron)
    logo = se.makePicture(logo)
    canvas = se.makePicture(canvas)
    
    copy(thor,canvas,0,0)
    copy(cap,canvas,200,0)
    copy(hulk,canvas,400,0)
    copy(iron,canvas,600,0)
    copy2(logo,canvas,640,310)
    
    canvas.show()
```
