```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, label='sin')
plt.plot(x,y2, label='cos', linestyle='--')
plt.title('sin&cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
